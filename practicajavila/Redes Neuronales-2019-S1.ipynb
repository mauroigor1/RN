# Conjuntos de entrenamiento, validación y testeo. 
En el contexto del aprendizaje supervisado tenemos la capacidad de evaluar y cuantificar, bajo ciertas métricas, el desempeño de nuestro modelo. Por supuesto que esto aplica directamente a redes neuronales. 

En estos problemas nuestro objetivo es entrenar un modelo que sea capaz de predecir/clasificar algún fenómeno en función de cierta información. El proceso se separa en dos: 

1. ***Etapa de Entrenamiento***: usualemente el 70% u 80% de nuestros datos totales se destina a la etapa de entrenamiento. Estos datos son nuevamente separados en dos conjuntos **disjuntos**:
  * ***Conjunto de Entrenamiento:*** corresponde al conjunto de datos con el que se entrenará el modelo.           
  * ***Conjunto de Validación:*** corresponde a un pequeño conjunto de datos que, en cada época del entrenamiento se evalúa sobre el modelo. Esto nos dará una idea cercana sobre qué performance tendrá el modelo sobre los datos de entrenamiento. También nos servirá para analizar si nuestro modelo está aprendiendo bien, se está sobre ajustando (overfitting), o está aprendiendo de muy mala manera (underfitting). _Más adelante veremos cómo utilizar los resultados de este conjunto de validación_. 
  
2. ***Etapa de Testeo***: los datos restantes luego de elegir los de entrenamiento (30% o 20%). Son utilizados para evaluar el desempeño del modelo entrenado sobre datos que jamás ha visto antes. Si asumimos que nuestros datos son consistentes y los datos de entrenamiento contienen suficientes datos para abarcar la variabilidad de los datos existentes, entonces deberíamos obtener resultados de desempeño similares a los obtenidos en el entrenamiento

## Funciones de activación 
Algunas de las funciones de activación más usuales son: 

### 1) Función Sigmoidal
\begin{equation}
  g(z) = \dfrac{1}{1+e^{-z}}
\end{equation}

### 2) Función Tangente Hiperbólica
\begin{equation}
  g(z) = \dfrac{e^z - e^{-z}}{e^z+e^{-z}}
\end{equation}

### 3) Función ReLU 
\begin{equation}
  g(z) = \max(0,z)
\end{equation}

### 4) Funcion Leaky ReLU
\begin{equation}
  g(z) = \max(\varepsilon z,z) , \varepsilon << 1
\end{equation}

***Graficaremos algunas funciones de activación***

```python
import matplotlib.pyplot as plt
import numpy as np
#%matplotlib inline
%matplotlib notebook


# Definir la función sigmoidal
def sigmoid(z):
    return(1/(1+ np.exp(1)**-z))

# Función Tangente Hiperbólica
def tanh(z):
    return (np.exp(1) - np.exp(1)**-z) / (np.exp(1) + np.exp(1)**-z) 

# Función ReLU
def relu(z): 
    return max(0.0,z)

# Función Leaky ReLU
def leakyrelu(z, eps=0.1):
    'Por defecto se definió un eps = 0.01'
    return max(eps*z,z)

# Vectorizar las funciones 
Sigmoid = np.vectorize(sigmoid)
Tanh    = np.vectorize(tanh)
Relu    = np.vectorize(relu)
LeakyRelu = np.vectorize(leakyrelu) 

# Generar un vector desde -10 hasta 10 con 200 elementos equiespaciados 
z = np.linspace(start=-10,stop=10,num=200) 

## Graficamos las funciones
# Sigmoidal
plt.subplot(221)
plt.plot(z, Sigmoid(z))
plt.title('Función Sigmoidal')
plt.ylabel('sigmoid')
plt.xlabel('z')
plt.grid(True)

# Tangente Hiperbólica
plt.subplot(222)
plt.plot(z, Tanh(z))
plt.title('Función tanh')
plt.ylabel('tanh')
plt.xlabel('z')
plt.grid(True)

# ReLU
plt.subplot(223)
plt.plot(z, Relu(z))
plt.title('Función ReLU')
plt.ylabel('tanh')
plt.xlabel('z')
plt.grid(True)


# Leaky ReLU
plt.subplot(224)
plt.plot(z, LeakyRelu(z))
plt.title('Función Leaky ReLU')
plt.ylabel('tanh')
plt.xlabel('z')
plt.grid(True)

plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.5,
                    wspace=0.5)
```

Una ***buena práctica*** es reescalar los datos, de modo de tener a todos en un rango similar. Ejemplos de esto son la estandarización al intervalo $[0,1]$, $[-1,1]$ o una normalización del tipo $\dfrac{x-\mu}{\sigma}$, donde $\mu$ es el promedio de las observaciones de una variable específica y $\sigma$ su desviación estándar.

# Escalamientos clásicos para un dato 
Supongamos que un dato $x$ de una muestra $X = {x_1,...,x_n}$ lo queremos reescalar a un valor $z$. Conocemos $\mu$ (el promedio de la muestra $X$), $\sigma$ (su desviación estándar), $\max$ y $\min$ de $X$.
### Al intervalo $[0,1]$  
Note que aquí necesitamos encontrar una función $f: Dom(f) \subset \mathbb{R} \rightarrow [0,1]$
donde $Dom(f) = [\min, \max]$.

Teniendo lo anterior en consideración, es sencillo observar que la función buscada para reescalar nuestros datos al intervalo $[0,1]$ está dada por:

$$f(x) = \dfrac{x - \max}{\max - \min} $$


### Al intervalo $[-1,1]$ 
Ahora, si multiplicamos la función anterior por $2$, obtendremos resultados entre $[0,2]$, luego basta trasladar la función en $1$ para obtener resultados en el intervalo $[-1,1]$.
$$f = 2\dfrac{x - \max}{\max - \min} - 1$$

### A un intervalo $[a,b]$ cualquiera
Y ahora puede notar que en realidad el caso más general, donde $Rec(f)=[a,b]$ nos lleva a definir un escalamiento de la forma: 
$f(x) = (b-a)\dfrac{x - \max}{\max - \min} + a$

### Normalizar como una $\mathcal{N}(0,1)$
Otro escalamiento clásico es normalizar los datos, es decir, obligarlos a que el promedio de los datos transformados sea igual a 0, y que su desviación estándar sea 1. Esto hará que la mayoría de los datos viva alrededor del 0, variando entre $-1$ y $1$ (pero ojo, la mayoría, no todos).

$$z = \dfrac{x-\mu}{\sigma}$$

A continuación puede ver un esquema de cómo se vería una distribución normal de datos. [Puede leer más acá](https://es.wikipedia.org/wiki/Distribuci%C3%B3n_normal)

```python
Image(url = 'https://upload.wikimedia.org/wikipedia/commons/thumb/3/3a/Standard_deviation_diagram_micro.svg/1280px-Standard_deviation_diagram_micro.svg.png')
```

***Ahora que tenemos nuestros datos escalados, debemos elegir nuestros conjuntos de entrenamiento y testeo para luego entrenar una red***

```python
# Separar el conjunto total de datos en entrenamiento y testeo. 
import numpy as np

# Elegimos una cantidad
porcTrain = 0.7 # Porcentaje de los datos a considerar en el entrenamiento
Ntrain = int(Data.shape[0]//(1/porcTrain)) # Cantidad de elementos en el conjunto de entrenamiento
lista  = Data.index.tolist() # La lista de los indices

# Seleccionar lista de indices de Data para el entrenamiento
sampleTrain = np.random.choice( 
                  a       = lista,         # Lista sobre la cual elegiré mi muestra
                  size    = Ntrain,        # Cantidad de elementos que quiero retornar
                  replace = False       # Con reemplazo o no
                )

# Seleccionar los indices restantes para dejar en el conjunto de testeo
sampleTest = [ind for ind in set(lista) - set(sampleTrain)]
```

```python
print('Cantidad de elementos en el conjunto de entrenamiento: ' + str(len(sampleTrain)))
print('En el conjunto de test: ' + str(len(sampleTest)))
print('Entrenamiento + Testeo = ' + str(len(sampleTrain) + len(sampleTest)))
```

```python
colsInput = ['Female', 'Male', 'Age', 'Spending Score (1-100)']
InputTrain  = Data.loc[sampleTrain, colsInput]
OutputTrain = Data.loc[sampleTrain, 'Annual Income (k$)']
```

```python
# Importar librería keras
from keras.models import Sequential
from keras.layers import Dense

# Crear Modelo secuencial
model = Sequential()

# Agregamos capas a nuestro modelo
model.add(Dense(10, input_dim = InputTrain.shape[1], activation = 'relu')) # Primero la capa de entrada
model.add(Dense(12,activation = 'relu')) # Agregando capas
model.add(Dense(1, activation = 'softmax')) #¿Es correcta la función de salida? Explique por qué. 

model.summary()
```

```python
# Compilar el modelo
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['mean_squared_error'])
```

```python
# Entrenar el modelo
history = model.fit(InputTrain, OutputTrain, validation_split=0.3, epochs=10, batch_size=10)
```

Otra situación que podemos encontrar es el **Overfitting**, situación en la cual el modelo se sobre ajusta a los datos de entrenamiento, es decir, los memoriza, y no es capaz de generalizar a datos nuevos, que serían los datos de testeo.

***¿Hay alguna manera de usar el overfitting a mi favor?***
El usar demasiadas épocas tiene la desventaja de que es altamente probable entrar en overfitting, pues los datos pasarán tantas veces por la red que esta tenderá a memorizarlos, pero si nuestro modelo se fue entrenando bien hasta cierta época, podríamos querer quedarnos con el mejor modelo que se haya entrenado (luego de cada época tenemos un modelo distinto, pues son nuevos pesos los que se actualizan). 

Afortunadamente tenemos una forma sencilla de hacer esto. Es una funcionalidad de keras llamada **checkpoint**, la cual consiste en guardar los pesos después de cada época, siempre y cuando el valor loss_val (pérdida en el conjunto de validación) decrezca. Esto se aplica previo a entrenar nuestro modelo. Lo veremos en un ejemplo más adelante.
